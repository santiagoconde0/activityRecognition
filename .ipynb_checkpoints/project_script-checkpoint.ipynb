# Code: 
Libreries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import mode
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels
from sklearn.preprocessing import StandardScaler
```

Paths and interactions

```python
main_path = 'C:\\Users\\s7277028\\Documents\\ARSSD\\project'
users = range(1,9)
trials = range(1,5)
lbs = range(0,6)
```

Preprocessing

```python
all_data = pd.DataFrame()

#Read data for all users, trials and labels
for user in users:
    for trial in trials:
        for lb in lbs:
            filePath = main_path + "\\data\\User"+str(user)+"\\Trial"+str(trial)+"\\L"+str(lb)+".csv"
            sadl1 = pd.read_csv(filePath, sep=',', header=0)
            data = sadl1.iloc[:, :79]
            
            # Select the sensors
            data=data[data.columns[np.r_[0:1,21:24,25:28]]]
            labels = sadl1.iloc[:,78]

            #find and remove rows with all nulls
            idx=data.index[data.isnull().all(1)]
            data = data[~data.index.isin(idx)]
            labels = labels[~labels.index.isin(idx)]
            labels.value_counts()

            #fill missing values
            data = data.fillna(method='ffill',axis=1)
            data['user'] = user
            data['trial'] = trial
            
            #The data is filtered
            columns = data.columns[~data.columns.isin(['user', 'trial', 'loggingTime'])]
            filtered_data = data[columns].rolling(5).median()
            filtered_data = filtered_data.fillna(method='ffill',axis=1)
            filtered_data['loggingTime'] = data.loggingTime
            
            #segmentation
            filtered_data['time']=pd.to_datetime(data.loggingTime, unit='s')
            filtered_data.index=filtered_data.time
            #Overlap 50%
            keep = filtered_data.time.dt.microsecond/1000 %500
            keep = keep - keep.shift() < 0
            
            #Window of 1 second
            means = filtered_data[columns].rolling('1S').mean()[keep]
            means.columns = [str(col) + '_mean' for col in means.columns]
            variances = filtered_data[columns].rolling('1S').var()[keep]
            variances.columns = [str(col) + '_var' for col in variances.columns]

            labels.index = filtered_data.time
            mode_labels = labels.rolling('1S').apply(lambda x:mode(x)[0])[keep]

            #all features
            all_features = pd.concat([means, variances], axis=1)
            all_features['label'] = mode_labels 
            all_features['user'] = user
            all_features['trial'] = trial
            all_data = pd.concat([all_data, all_features])
```

`fucntion to plot the confusion matrix`

```python
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Greens):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        
    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

# Basic Chain

*To train the models we are going to use the classifier **RandomForestClassifier**, to avoid the change of the results because of the random state, we are going to set a seed of 0.*

### Split data by Trials: 

*The trials 1 and 4 are going to be the data for the training and the 2 and 3 the test*

```python
# Split the data
train_trials = ['1','4']
train_data = all_data[all_data.trial.isin(train_trials)]
train_data = train_data.fillna(train_data.mean())
test_data = all_data[~all_data.trial.isin(train_trials)]
test_data = test_data.fillna(test_data.mean())

# Dedife train data
X_train = train_data.loc[:,~train_data.columns.isin(['user', 'trial','label'])]
y_train = train_data.loc[:,train_data.columns.isin(['label'])]

#Define test data
X_test = test_data.loc[:,~test_data.columns.isin(['user', 'trial','label'])]
y_test = test_data.loc[:,test_data.columns.isin(['label'])]

# Train the model
classifier =RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1, random_state=0)
classifier.fit(X_train, y_train)

# Test the model
score = classifier.score(X_test, y_test)
print("Accuracy: ", score)
y_pred = classifier.predict(X_test)
print(classification_report(y_test, y_pred))

np.set_printoptions(precision=2)
class_names = np.array(['Other', 'Stand up' , 'Grab' , 'Point', 'Shoot', 'Load'])

plot_confusion_matrix(y_test, y_pred, classes=class_names,
                      title='Confusion matrix, without normalization')

plot_confusion_matrix(y_test, y_pred, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```

### Split data by Users: 

*The users 1, 2 , 3 and 6 are going to be the data for the training and the 7 and 8 for test*

```python
# Split the data
train_users = ['1','2','3','6']
train_data = all_data[all_data.user.isin(train_users)]
train_data = train_data.fillna(train_data.mean())
test_data = all_data[~all_data.user.isin(train_users)]
test_data = test_data.fillna(test_data.mean())

# Dedife train data
X_train = train_data.loc[:,~train_data.columns.isin(['user', 'trial','label'])]
y_train = train_data.loc[:,train_data.columns.isin(['label'])]

#Define test data
X_test = test_data.loc[:,~test_data.columns.isin(['user', 'trial','label'])]
y_test = test_data.loc[:,test_data.columns.isin(['label'])]

# Train the model
classifier =RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1, random_state=0)
classifier.fit(X_train, y_train)

# Test the model
score = classifier.score(X_test, y_test)
print("Accuracy: ", score)
y_pred = classifier.predict(X_test)
print(classification_report(y_test, y_pred))

np.set_printoptions(precision=2)
class_names = np.array(['Other', 'Stand up' , 'Grab' , 'Point', 'Shoot', 'Load'])

plot_confusion_matrix(y_test, y_pred, classes=class_names,
                      title='Confusion matrix, without normalization')

plot_confusion_matrix(y_test, y_pred, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```

## Transfer learning

*To try to improve the behavior of the "Point" class we are going to apply the method of transfer learning to it. The transfer learning will be applied to the labels, using as the original task the labels differents to "Point" and the transfer task will be the "Point".*

Split the data by trials and define the "Original task" as the "other" class

```python
# Split the data in Original and transfer data
train_trials = ['1','4']
train_data = all_data[all_data.trial.isin(train_trials)]
original_task = train_data[train_data.label ==0]
transfer_data = train_data[train_data.label == 3]

# Scale the original data 
scaler = StandardScaler()
scaler.fit(original_task)
colunmns = original_task.columns
original_task = scaler.transform(original_task)
original_task = pd.DataFrame(original_task, columns=colunmns)

# Get the mean and std of the transfer task
scaler = StandardScaler()
scaler.fit(transfer_data)
transfer_mean = scaler.mean_
transfer_std = scaler.scale_

# Multiply the std and add the mean to the scale data
for name_column, std, mean in zip(original_task.columns.values,transfer_std,transfer_mean):
    serie = original_task[name_column]
    serie = serie*std + mean
    original_task[name_column] = serie
    
# original_task.head()

# Add the original users, trials and the new label
for i in range(len(original_task.user)):
    original_task.user[i] = train_data[train_data.label == 0].user[i]
    original_task.trial[i] = train_data[train_data.label == 0].trial[i]
    original_task.label[i] = 3
    
# Concat the new dataframe with the olds
tl_data = pd.DataFrame()
tl_data = pd.concat([tl_data, train_data[train_data.label != 3]])
tl_data = pd.concat([tl_data, original_task])
```

Train the model with the transfer learning changes

```python
# Split the data for traing and test
train_trials = ['1','4']
train_data = tl_data.fillna(test_data.mean())
test_data = all_data[~all_data.trial.isin(train_trials)]
test_data = test_data.fillna(test_data.mean())

# Dedife train data
X_train = train_data.loc[:,~train_data.columns.isin(['user', 'trial','label'])]
y_train = train_data.loc[:,train_data.columns.isin(['label'])]

#Define test data
X_test = test_data.loc[:,~test_data.columns.isin(['user', 'trial','label'])]
y_test = test_data.loc[:,test_data.columns.isin(['label'])]

# Train the model
classifier =RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1, random_state=0)
classifier.fit(X_train, y_train)

# Test the model
score = classifier.score(X_test, y_test)
print("Accuracy: ", score)
y_pred = classifier.predict(X_test)
print(classification_report(y_test, y_pred))

np.set_printoptions(precision=2)
class_names = np.array(['Other', 'Stand up' , 'Grab' , 'Point', 'Shoot', 'Load'])

plot_confusion_matrix(y_test, y_pred, classes=class_names,
                      title='Confusion matrix, without normalization')

plot_confusion_matrix(y_test, y_pred, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```
